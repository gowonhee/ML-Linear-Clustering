```python
# 이직률 계산해보기!
# ~~률 -> 확인하고자 하는 데이터 개수 / 전체 데이터 개수
(df["Attrition"] == "Yes").sum() / len(df)
# 이직률은 약 16012% 정도로 확인!
```

```python
df["Attrition"]
```

```python
# 학습 및 집계를 하기위해 수치데이터로 변경
# 이직 = 
# np.where(조건식, True인 경우 변경할 값, False인 경우 변경할 값)
df["Attrition"] = np.where(df["Attrition"] == "Yes", 1, 0) # 오렌지 Formula와 비슷
```

```python
df["Attrition"].value_counts()
```

```python
# 연력 분석
# 연령별로 범주화(Bining) 이후, 이직률 파악해보자!
# 총 3개의 범주로 Bining(범주화)
df["Age_gp"] = np.where(df["Age"] <= 30, "30 이하",
        np.where(df["Age"] <= 40, "31~40대", # 나이가 31~40인 경우, "31~40대"로 그룹화
                "40대 이상")) # 40을 초과하는 경우, "40대 이상"
```

```python
# 연령별 이직률 현황
df_gp = df.groupby("Age_gp")["Attrition"].agg(["count", "sum"])
# count : 각 그룹에 속한 총 직원 수
# sum : 이직한 직원의 수

df_gp
```

```python
# 확인하고자 하는 데이터 개수 / 전체 데이터 개수
df_gp["ratio"] = round(df_gp['sum'] / df_gp["count"] * 100, 1)
df_gp
# 나이가 어릴수록 이직 비율이 높아짐 -> 나이라는 특성이 이직률에 어느정도 관여를 한다!
```

```python
# 성별 이직률 현황
df["Gender"] = np.where(df["Gender"] == "Female" , "Female", "Male")
df_gen_gp = df.groupby("Gender")["Attrition"].agg(["count", "sum"])
df_gen_gp["ratio"] = round(df_gen_gp['sum'] / df_gen_gp["count"] * 100, 1)
display(df_gen_gp)

# 부서별 이직률 현황
depart_gp = df.groupby("Department")["Attrition"].agg(["count", "sum"])
depart_gp["ratio"] = round(depart_gp['sum'] / depart_gp["count"] * 100, 1)
display(depart_gp)

# 남성의 이직률이 조금 더 높음
# Sales 부서의 이직률이 가장 높음
```

#### 가설1) 업무 만족도는 높으나, 인간관계로 인한 이직률이 높을 것이다!

```python
# JobSatisfaction : 업무 만족도(숫자)
# RelationshipSatisfaction : 인간관계 만족도 (숫자가 클수록 만족)
df[["JobSatisfaction", "RelationshipSatisfaction", "Attrition"]].head()
```

```python
# 업무 만족도 별 이직률 현황
jobsatis_gp = df.groupby("JobSatisfaction")["Attrition"].agg(["count","sum"])
jobsatis_gp["ratio"] = round(jobsatis_gp['sum'] / jobsatis_gp["count"] * 100, 1)
display(jobsatis_gp)
# 업무만족도가 낮을 수록 이직률이 높다!
```

```python
# 업무만족도, 인간관계만족도 별 이직률 현황
job_relationship_gp = df.groupby(["JobSatisfaction", "RelationshipSatisfaction"])["Attrition"].agg(["count","sum"])
job_relationship_gp["ratio"] = round(job_relationship_gp['sum'] / job_relationship_gp["count"] * 100, 1)
display(job_relationship_gp)

# 업무만족도가 높은 직원은 인간관계에 따라 이직률에 영향을 덜 받는 것으로 관측됨
# 업무만족도가 낮은 직원은 인간관계가 나쁠수록 이직률이 증가하는 경향을 보임
```

가설 2) 근속년수 대비 같은 업무를 했던 비중이 높다면, 이직률이 높지 않을까?

```python
df[['YearsAtCompany','YearsInCurrentRole']]
# YearsAtCompany : 근속년수 (입사 이후 현재까지 근무한 기간)
# YearsInCurrentRole : 현재 역할에서 근무한 기간
```

```python
# 근속년수 대비 한가지 업에 대한 비중
df["Role_Company"] = df["YearsInCurrentRole"] / df["YearsAtCompany"] 
df["Role_Company"].fillna(0, inplace = True) # 0/0은 존재할 수 없음! -> 결측치라서 대체함!
```

```python
df["Role_Company"].isnull().sum()
```

```python
sns.distplot(df["Role_Company"]);
# 히스토그램 : 데이터의 빈도 분포를 막대로 표현한 그래프
# 커널 밀도 추정(KDE, Kernel Density, Estimate) : 히스토그램을 부드럽게 표현한 곡선형 그래프

# Role_Company : 근속년수 대비 현재 직무의 근무한 비중
# 0에 가까운 피크 -> 많은 직원들이 현재 직무에서 근무한 기간이 전체 회사 근무 기간에 비해 매우 짧다!
```

```python

# 근속년수 대비 한가지 업무 비율(Role_Company)을 범주화하여, 이직률의 비율을 어떠한지 그룹테이블 만들어보기!
# 범주 : 0.3이하, 0.3~0.6, 0.6초과
df["Role_Company_gp"] = np.where(df["Role_Company"] <= 0.3, "0.3이하",
         np.where(df["Role_Company"] <= 0.6, "0.3~0.6", "0.6초과"))
```

```python
# 이직률 확인
role_company_gp = df.groupby("Role_Company_gp")["Attrition"].agg(["count","sum"])
role_company_gp["ratio"] = round(role_company_gp["sum"] / role_company_gp["count"] * 100,1)
role_company_gp
# 근속년수 대비 같은 업무를 오래 한 경우 이직률이 낮다!
```
