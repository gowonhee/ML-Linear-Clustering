```python
# 라벨 인코딩
from sklearn.preprocessing import LabelEncoder
label_columns = [
    "Displaced", 
    "Educational special needs", 
    "Debtor", 
    "Tuition fees up to date", 
    "Gender", 
    "Scholarship holder", 
    "International", 
    "Target", 
    "Daytime/evening attendance"
]
label_enc = LabelEncoder()

for column in label_columns:
    df[column] = label_enc.fit_transform(df[column])

df
```

```python
# 원핫 인코딩
ont_hot_columns = [
    "Marital status",
    "Course",
    "Previous qualification",
    "Nationality",
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation"
]
df_one_hot = pd.get_dummies(df, columns=ont_hot_columns, dtype=np.int64)
```

```python
df = df_one_hot
```

```python
df_one_hot["Target"].value_counts()
# 0 = drop out(중퇴)
# 1 = enrolled (재학중)
# 2 = graduate(졸업)
```

```python
# 수업료 납부 여부 상태 (수치 데이터) 
df["Tuition fees up to date"].value_counts()
```

```python
# 수업료 납부를 안한 사람(0)일수록 target(0)"중퇴"했을 확률이 높다.

df_gp = df.groupby("Target")["Tuition fees up to date"].agg(["count", "sum"])
# count : 총 학생 수
# sum : 

df_gp
```

```python
df_gp["ratio"] = round(df_gp['sum'] / df_gp["count"] * 100, 1)
df_gp
```

```python
df["target_gp"] = np.where(df["Target"] == 0, "중퇴",
         np.where(df["Target"] == 1, "재학 중", "졸업"))
```
